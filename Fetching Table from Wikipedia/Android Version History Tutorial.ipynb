## Data Scrapping using Beautiful Soup
- Import Beautiful Soup
-Making a GET request to fetch the data
-Parse the HTML

```python
#STEP-1 Import Beautiful Soup
#STEP-2: Open the desired webpage from Python
# To Open a URL we have to import the urllib.request module and call the urlopen method
#urllib.request is a Python module for fetching URLs (Uniform Resource Locators). 
#It offers a very simple interface, in the form of the urlopen function.
#This is capable of fetching URLs using a variety of different protocols
from urllib.request import urlopen
scrapping_website="https://en.wikipedia.org/wiki/Android_version_history" #Specifies which URL open
website_data=urlopen(scrapping_website) # Fetches the data from given webpage
print(type(website_data))
# Now to get the HTML data we need to 'read' the fetched response
desired_html=website_data.read()
#After receving the HTML data we should close the connection
website_data.close()
```

### Parsing the Data using Beautiful Soup

```python
#First we have import BeautifulSoup
from bs4 import BeautifulSoup as soup
# Here BeautifulSoup is a class and soup will help us creating objects of BeautifulSoup class
req_doc=soup(desired_html,"html.parser")
req_doc.h1 #This method gives us the 1st h1 as output
```

```python
req_doc.findAll('table',{}) #Gives us all the tables and the empty dict. is there as no class is selected
```

```python
tables=req_doc.findAll('table',{'class':'wikitable'}) #Finds all the tables with class as wikitables
print(len(tables))
```

```python
version_history=tables[0]
print(version_history)
```

```python
first_row=version_history.findAll('th',{}) #Scrapping the column names of the table
print((first_row))
```

```python
print(first_row[0].text)  #Getting the name of the 1st column of the table
```

```python
column_heading=[ x.text[:-1] for x in first_row] 
print(column_heading)
```

```python
data_rows=version_history.findAll('tr')[1:]
print(len(data_rows))
```

```python
#Extracting Data from the 1st row with data
first_data_row=data_rows[0].findAll('td',{})
print(first_data_row)
```

```python
inital_row_entries=[x.text[:-1] for x in first_data_row]
print(inital_row_entries)
```

```python
table_data=[] #This list will contain the data of all the rows
for x in data_rows: #x is the current row
    current_row=[]  #This list will store all the table data of a particular row one by one
    ans=x.findAll("td",{}) #All the table data of 1 row is extracted as a whole
    for data in (ans):#The data data is accessed one by one item
        current_row.append(data.text[:-1]) # All the singular table data is stored as the elements of 
                                            #current row
    table_data.append(current_row)#Rows as a whole is stored in this list
        
```

```python
print(table_data)
```

## Writing CSV file

```python
filename="android_history.csv"
with open(filename,"w",encoding='utf-8') as f:
    column_name_string=','.join(column_heading)
    column_name_string+="\n"
    f.write(column_name_string)
    
    for row in table_data[:-1]:
        row_string=""
        for w in row:
            w=w.replace(',','')
            row_string+=w+','
        row_string=row_string[:-1]
        row_string+="\n"
        f.write(row_string)
```
