```python
from urllib.request import urlopen
url="https://en.wikipedia.org/wiki/Call_of_Duty"
cod_url=urlopen(url)
cod_data=cod_url.read()
cod_url.close()
```

```python
from bs4 import BeautifulSoup as soup
actual_html=soup(cod_data,'html.parser')
actual_html.h1
```

```python
cod_table=actual_html.findAll('table',{'class':'wikitable'})
print(len(cod_table))
```

```python
table=cod_table[1]
first_row=table.findAll('th',{})
column_list=[]
for i in first_row:
    column_list.append(i.text[:-1])
column_list
```

```python
rest_rows=table.findAll('tr',{})
rest_rows=rest_rows[1:]
data_rows=[]

for rows in rest_rows:
    current_row=[]
    req_ans=rows.findAll('td',{})
    for val in (req_ans):
        current_row.append(val.text[:-1])
    data_rows.append(current_row)
data_rows
        
        
```

```python
filename="cod_history.csv"
with open(filename,"w",encoding="utf-8") as f:
    column_name_string=','.join(column_list)
    column_name_string+="\n"
    f.write(column_name_string)
    
    for row in data_rows:
        row_string=""
        for w in row:
            w=w.replace(',',' ')
            row_string+=w+','
        row_string=row_string[:-1]
        row_string+="\n"
        f.write(row_string)
            
```
